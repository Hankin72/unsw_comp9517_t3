```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# read wagon.jpg
wagon = cv2.imread("wagon.jpg")
wagon_rgb = cv2.cvtColor(wagon, cv2.COLOR_BGR2RGB)
plt.imshow(wagon_rgb),plt.xticks([]), plt.yticks([])
plt.show()

```

```python
# read wheel.png
wheel = cv2.imread("wheel.png")
wheel_rgb = cv2.cvtColor(wheel, cv2.COLOR_BGR2RGB)
plt.imshow(wheel_rgb),plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# read wheel.png
toy = cv2.imread("toy.png")
toy_rgb = cv2.cvtColor(toy, cv2.COLOR_BGR2RGB)
plt.imshow(toy_rgb),plt.xticks([]), plt.yticks([])
plt.show()
```

```python
def get_keypoint_image(img):
    #create an extractor inclduing hessian_threshold value
    surf = cv2.xfeatures2d_SURF.create(500)
    #compute its SURF features, key_query = keypoints

    key,desc = surf.detectAndCompute(img, None )
    # Display the image with fewer keypoints

    surf_img = cv2.drawKeypoints(img,key, np.array([]))

    return surf, surf_img, key,desc
    
```

```python
#display wagon keypoints dectected iamge
surf1, surf_wagon, key_query,desc_query  = get_keypoint_image(wagon_rgb)
plt.imshow(surf_wagon)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
##display wheel keypoints dectected iamge
surf2, surf_wheel, key_2,desc_2  = get_keypoint_image(wheel_rgb)
plt.imshow(surf_wheel)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
##display toy keypoints dectected iamge
surf3, surf_toy, key_toy,desc_toy  = get_keypoint_image(toy_rgb)
plt.imshow(surf_toy)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# mathch wagon and wheel
# def sift_matches(img1,img2, )
xy = cv2.BFMatcher()
wagon_wheel_matches = xy.knnMatch(desc_query, desc_2, k=2)
# matches = sorted(matches, key=lambda x: x.distance)
wagon_2_wheel = [[m] for m,n in wagon_wheel_matches if m.distance < 0.5 * n.distance]
output_1 = cv2.drawMatchesKnn(surf_wagon, key_query, surf_wheel, key_2, wagon_2_wheel, None, flags=2)
plt.imshow(output_1)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# mathch wagon and wheel
# def sift_matches(img1,img2, )
mn = cv2.BFMatcher()
wagon_toy_matches = mn.knnMatch(desc_query, desc_toy, k=2)
# matches = sorted(matches, key=lambda x: x.distance)

wagon_2_toy = [[m] for m,n in wagon_toy_matches if m.distance < 0.5 * n.distance]
output2 = cv2.drawMatchesKnn(surf_wagon, key_query, surf_toy, key_toy, wagon_2_toy, None, flags=2)
plt.imshow(output2)
plt.xticks([]), plt.yticks([])
plt.show()
```
