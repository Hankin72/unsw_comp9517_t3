```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import copy

# read an image in grayscale
# cv2.imread()
image5 = cv2.imread("imageQ5.png",0)
plt.imshow(image5, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([]) # to hide tick values on x and y
plt.show()
```

```python
# Step 1. Using Low-pass filter--median dilter to remove salt-and-pepper noise
median_image5 = cv2.medianBlur(image5,5)
plt.imshow(median_image5, cmap='gray', vmin=0, vmax=255)
plt.xticks([]), plt.yticks([]) 
plt.show()
```

```python
# Step 2. Histogram Equalization 
hist = cv2.calcHist([median_image5],[0], None, [256],[0,256]) 

plt.subplot(2,1,1),plt.imshow(median_image5,cmap = 'gray', vmin=0, vmax=255)

plt.subplot(2,1,2)
plt.hist(median_image5.ravel(),256,[0,256])
plt.show()
```

```python
#  histogram equalize by opencv function cv2.equalizeHist()

equ_img5 = cv2.equalizeHist(median_image5)
hist = cv2.calcHist([equ_img5],[0], None, [256],[0,256]) 

plt.subplot(2,1,1),plt.imshow(equ_img5,cmap = 'gray', vmin=0, vmax=255)

# show the Histogram after equalization
plt.subplot(2,1,2)
plt.hist(equ_img5.ravel(),256,[0,256])
plt.show()
```

```python
# Step3, sharpening, using sobel and nsharpen masking

# sobelx8u = cv2.Sobel(equ_img5, -1, 1,0)
# # sobelx8U = cv2.Sobel(equ_img5, -1, 1,0,ksize=5)
# # #这里的-1，表示的是保持和原图一样的数据类型
# sobelx64f = cv2.Sobel(equ_img5, cv2.CV_64F, 1,0,ksize=5)
# abs_sobel64f =np.absolute(sobelx64f)
# sobel_8u = np.uint8(abs_sobel64f)

sobel_x = cv2.filter2D(equ_img5, -1,1,0)

gaussian_img5 = cv2.GaussianBlur(equ_img5, (5,5), 0)

mask = cv2.subtract(equ_img5, gaussian_img5)
img5_output = cv2.addWeighted(sobel_x, 1, mask, 1.35, 0)


plt.imshow(img5_output,cmap = 'gray', vmin=0,vmax=255)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# Step4: contrast stretching 

# print(np.min(img5_output))
# print(np.max(img5_output))

a=0
b=255
c=np.min(img5_output)
d=np.max(img5_output)       
img5_final_output = (img5_output - c)*((b-a)/(d-c)) + a 

plt.imshow(img5_final_output, cmap='gray',vmin=0,vmax=255)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# save output
cv2.imwrite('q5_final_output.png',img5_final_output)
```
